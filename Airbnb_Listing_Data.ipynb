```python
import pandas as pd

listings = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Airbnb Data/Listings.csv",encoding = "ISO-8859-1",low_memory=False)
```

```python
listings.head()
```

```python
listings.info()
```

```python
listings["host_since"] = pd.to_datetime(listings["host_since"])
```

```python
listings.info()
```

```python
paris_listings = listings.query("city =='Paris'").loc[:,["host_since","neighbourhood","city","accommodates","price"]]
paris_listings.info()
```

```python
paris_listings.isna().sum()
```

```python
paris_listings.describe()
```

```python
paris_listings.query("accommodates==0 and price==0").count()
```

```python
paris_listings_neighbourhood = paris_listings.groupby("neighbourhood").agg({"price":"mean"}).sort_values("price")
paris_listings_neighbourhood.head()
```

```python
paris_listings_accommodats = paris_listings.query("neighbourhood == 'Elysee'").groupby("accommodates").agg({"price":"mean"}).sort_values("price")
paris_listings_accommodats.head()
```

```python
paris_listings_over_time = (
  paris_listings.set_index("host_since").resample("Y").agg({
      "neighbourhood":"count",
      "price":"mean"
  }))
```

```python
paris_listings_over_time.head()
```

```python
paris_listings_neighbourhood=paris_listings_neighbourhood.reset_index()
```

```python

import plotly.express as px
fig = px.bar(paris_listings_neighbourhood, y="neighbourhood", x="price", title="Average price per neighbourhood in Paris")
fig.show()
```

```python
paris_listings_accommodats.head()
```

```python

paris_listings_accommodats["accommodates"] = paris_listings_accommodats["accommodates"].astype(str)
paris_listings_accommodats.info()
```

```python
import plotly.express as px

fig = px.bar(paris_listings_accommodats, y="accommodates", x="price", title="Average price per Accomodation Number")
fig.show()
```

```python
paris_listings_over_time.head()
```

```python

fig = px.line(paris_listings_over_time, y="neighbourhood", x=paris_listings_over_time.index, title="New AirBnb Hosts in Paris Over Time")
fig.show()
```

```python
fig = px.line(paris_listings_over_time, y="price", x=paris_listings_over_time.index, title="Average Price in Paris Over Time")
fig.show()
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots()

ax.plot(
    paris_listings_over_time.index,
    paris_listings_over_time["neighbourhood"],
    label="New Hosts",
    c='pink'
)

ax.set_ylabel("New Host")

ax2 = ax.twinx()

ax2.plot(
    paris_listings_over_time.index,
    paris_listings_over_time["price"],
    label="Average Price",
)
ax.set_title("2015 Regulations lead to fewer New hosts,higher prices")
```
